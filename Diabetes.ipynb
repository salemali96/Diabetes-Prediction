```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.cross_validation import train_test_split

# https://www.kaggle.com/brandao/diabetes/home

diabetes = pd.read_csv("/Users/alisalem/Downloads/diabetes/diabetic_data.csv",na_values="?")
```

```python
diabetes = pd.DataFrame(diabetes) 
diabetes.rename(columns={'glyburide-metformin': 'glyburide_metformin'}, inplace=True)
diabetes.rename(columns={'glipizide-metformin': 'glipizide_metformin'}, inplace=True)
diabetes.rename(columns={'glimepiride-pioglitazone': 'glimepiride_pioglitazone'}, inplace=True)
diabetes.rename(columns={'metformin-rosiglitazone': 'metformin_rosiglitazone'}, inplace=True)
diabetes.rename(columns={'metformin-pioglitazone': 'metformin_pioglitazone'}, inplace=True)

diabetes
```

```python
del diabetes["weight"]
del diabetes["medical_specialty"]
del diabetes["payer_code"]

diabetes.drop(['encounter_id','patient_nbr', 'admission_type_id', 'discharge_disposition_id','admission_source_id'], axis = 1, inplace = True)
```

```python
diabetes.race.mode()
```

```python
diabetes.race = diabetes.race.fillna(diabetes.race.mode()[0])
```

```python
diabetes.isnull().sum()
```

```python
diabetes.diag_1 = diabetes['diag_1'].astype(str)
for i in range(diabetes.shape[0]):
    if 'V' in diabetes.diag_1[i]:
        diabetes.diag_1[i] = diabetes.diag_1[i].replace("V","")
    if "E" in diabetes.diag_1[i]:
        diabetes.diag_1[i] = diabetes.diag_1[i].replace("E","")
    if 'nan' in diabetes.diag_1[i]:
        diabetes.diag_1[i] = diabetes.diag_1[i].replace("nan","")
diabetes.diag_1 = pd.to_numeric(diabetes.diag_1)
diabetes.diag_1 = diabetes.diag_1.fillna(diabetes.diag_1.mode()[0])
```

```python
diabetes.diag_2 = diabetes['diag_2'].astype(str)
for i in range(diabetes.shape[0]):
    if "V" in diabetes.diag_2[i]:
        diabetes.diag_2[i] = diabetes.diag_2[i].replace("V","")
    if "E" in diabetes.diag_2[i]:
        diabetes.diag_2[i] = diabetes.diag_2[i].replace("E","")
    if 'nan' in diabetes.diag_2[i]:
        diabetes.diag_2[i] = diabetes.diag_2[i].replace("nan","")
diabetes.diag_2 = pd.to_numeric(diabetes.diag_2)
diabetes.diag_2 = diabetes.diag_2.fillna(diabetes.diag_2.mode()[0])

```

```python
diabetes.diag_3 = diabetes['diag_3'].astype(str)
for i in range(diabetes.shape[0]):
    if "V" in diabetes.diag_3[i]:
        diabetes.diag_3[i] = diabetes.diag_3[i].replace("V","")
    if "E" in diabetes.diag_3[i]:
        diabetes.diag_3[i] = diabetes.diag_3[i].replace("E","")
    if 'nan' in diabetes.diag_3[i]:
        diabetes.diag_3[i] = diabetes.diag_3[i].replace("nan","")
diabetes.diag_3 = pd.to_numeric(diabetes.diag_3)
diabetes.diag_3 = diabetes.diag_3.fillna(diabetes.diag_3.mode()[0])
```

```python
diabetes['diag_1'] = diabetes['diag_1'].astype(int)
diabetes['diag_2'] = diabetes['diag_2'].astype(int)
diabetes['diag_3'] = diabetes['diag_3'].astype(int)
```

```python
diabetes.race = diabetes.race.astype('category')
diabetes.gender = diabetes.gender.astype('category')
diabetes.max_glu_serum = diabetes.max_glu_serum.astype('category')
diabetes.A1Cresult = diabetes.A1Cresult.astype('category')
diabetes.metformin = diabetes.metformin.astype('category')
diabetes.repaglinide = diabetes.repaglinide.astype('category')
diabetes.nateglinide = diabetes.nateglinide.astype('category')
diabetes.chlorpropamide = diabetes.chlorpropamide.astype('category')
diabetes.glimepiride = diabetes.glimepiride.astype('category')
diabetes.acetohexamide = diabetes.acetohexamide.astype('category')
diabetes.glipizide = diabetes.glipizide.astype('category')
diabetes.glyburide = diabetes.glyburide.astype('category')
diabetes.tolbutamide = diabetes.tolbutamide.astype('category')
diabetes.pioglitazone = diabetes.pioglitazone.astype('category')
diabetes.rosiglitazone = diabetes.rosiglitazone.astype('category')
diabetes.acarbose = diabetes.acarbose.astype('category')
diabetes.miglitol = diabetes.miglitol.astype('category')
diabetes.troglitazone = diabetes.troglitazone.astype('category')
diabetes.tolazamide = diabetes.tolazamide.astype('category')
diabetes.examide = diabetes.examide.astype('category')
diabetes.citoglipton = diabetes.citoglipton.astype('category')
diabetes.insulin = diabetes.insulin.astype('category')
diabetes.glyburide_metformin = diabetes.glyburide_metformin.astype('category')
diabetes.glipizide_metformin = diabetes.glipizide_metformin.astype('category')
diabetes.glimepiride_pioglitazone  = diabetes.glimepiride_pioglitazone.astype('category')
diabetes.metformin_rosiglitazone = diabetes.metformin_rosiglitazone.astype('category')
diabetes.metformin_pioglitazone = diabetes.metformin_pioglitazone.astype('category')
diabetes.change = diabetes.change.astype('category')
diabetes.diabetesMed = diabetes.diabetesMed.astype('category')
diabetes.readmitted = diabetes.readmitted.astype('category')


le=LabelEncoder()
le.fit(diabetes.race)
diabetes.race = le.transform(diabetes.race)

le.fit(diabetes.gender)
diabetes.gender = le.transform(diabetes.gender)

le.fit(diabetes.max_glu_serum)
diabetes.max_glu_serum = le.transform(diabetes.max_glu_serum)

le.fit(diabetes.A1Cresult)
diabetes.A1Cresult = le.transform(diabetes.A1Cresult)

le.fit(diabetes.metformin)
diabetes.metformin = le.transform(diabetes.metformin)

le.fit(diabetes.repaglinide)
diabetes.repaglinide = le.transform(diabetes.repaglinide)

le.fit(diabetes.nateglinide)
diabetes.nateglinide = le.transform(diabetes.nateglinide)

le.fit(diabetes.chlorpropamide)
diabetes.chlorpropamide = le.transform(diabetes.chlorpropamide)

le.fit(diabetes.glimepiride)
diabetes.glimepiride = le.transform(diabetes.glimepiride)

le.fit(diabetes.acetohexamide)
diabetes.acetohexamide = le.transform(diabetes.acetohexamide)

le.fit(diabetes.glipizide)
diabetes.glipizide = le.transform(diabetes.glipizide)

le.fit(diabetes.glyburide)
diabetes.glyburide = le.transform(diabetes.glyburide)

le.fit(diabetes.tolbutamide)
diabetes.tolbutamide = le.transform(diabetes.tolbutamide)

le.fit(diabetes.pioglitazone)
diabetes.pioglitazone = le.transform(diabetes.pioglitazone)

le.fit(diabetes.rosiglitazone)
diabetes.rosiglitazone = le.transform(diabetes.rosiglitazone)

le.fit(diabetes.acarbose)
diabetes.acarbose = le.transform(diabetes.acarbose)

le.fit(diabetes.miglitol)
diabetes.miglitol = le.transform(diabetes.miglitol)

le.fit(diabetes.troglitazone)
diabetes.troglitazone = le.transform(diabetes.troglitazone)

le.fit(diabetes.tolazamide)
diabetes.tolazamide = le.transform(diabetes.tolazamide)

le.fit(diabetes.examide)
diabetes.examide = le.transform(diabetes.examide)

le.fit(diabetes.citoglipton)
diabetes.citoglipton = le.transform(diabetes.citoglipton)

le.fit(diabetes.insulin)
diabetes.insulin = le.transform(diabetes.insulin)

le.fit(diabetes.glyburide_metformin)
diabetes.glyburide_metformin = le.transform(diabetes.glyburide_metformin)

le.fit(diabetes.glipizide_metformin)
diabetes.glipizide_metformin = le.transform(diabetes.glipizide_metformin)

le.fit(diabetes.glimepiride_pioglitazone )
diabetes.glimepiride_pioglitazone  = le.transform(diabetes.glimepiride_pioglitazone )

le.fit(diabetes.metformin_rosiglitazone)
diabetes.metformin_rosiglitazone = le.transform(diabetes.metformin_rosiglitazone)

le.fit(diabetes.metformin_pioglitazone )
diabetes.metformin_pioglitazone  = le.transform(diabetes.metformin_pioglitazone )

le.fit(diabetes.change)
diabetes.change = le.transform(diabetes.change)

le.fit(diabetes.diabetesMed)
diabetes.diabetesMed = le.transform(diabetes.diabetesMed)


diabetes.dtypes
```
